```python
import os

import h5py
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import sklearn
import scipy
#from pydiffmap.diffusion_map import DiffusionMap
import phate
from sklearn.random_projection import GaussianRandomProjection
from pydiffmap import diffusion_map as dm
import matplotlib.pyplot as plt

sns.set_context('notebook')
```

```python
os.chdir('..')
from fokker_plank_simdata import make_fp_explicit, make_fp_implicit
```

## High Drift

```python
# Generate the 1D Fokker-Planck data
np.random.seed(42)
p, samples, times = make_fp_implicit(dx=0.1, dt=0.05, sigma=0.5, k=100, time_steps=100, N=10, random_seed=42)

# = make_fp(sigma=0.5, k=100.0)

# Reshape samples for random projection
samples = samples.reshape(-1, 1)

# Random projection to 10D
transformer = GaussianRandomProjection(n_components=10)
data_10D = transformer.fit_transform(samples)
```

```python
# Create a diffusion map object with alpha = 0.5
mydmap = dm.DiffusionMap.from_sklearn(n_evecs=2, alpha=0.5, epsilon='bgh', k=64)
dmap = mydmap.fit_transform(data_10D)

# Plot the first non-trivial diffusion coordinates
plt.figure(figsize=(10, 6))
plt.scatter(dmap[:, 0], dmap[:, 1], c=times, cmap='Reds')
plt.colorbar(label='Simulation Time')
plt.xlabel('Diffusion Coordinate 1')
plt.ylabel('Diffusion Coordinate 2')
plt.title('Diffusion Map Embedding of 10D Projected Data')
plt.show()
```

```python
# Create a diffusion map object with alpha = 0.5
mydmap = dm.DiffusionMap.from_sklearn(n_evecs=2, alpha=1, epsilon='bgh', k=64)
dmap = mydmap.fit_transform(data_10D)

# Plot the first non-trivial diffusion coordinates
plt.figure(figsize=(10, 6))
plt.scatter(dmap[:, 0], dmap[:, 1], c=times, cmap='Reds')
plt.colorbar(label='Simulation Time')
plt.xlabel('Diffusion Coordinate 1')
plt.ylabel('Diffusion Coordinate 2')
plt.title('Diffusion Map Embedding of 10D Projected Data')
plt.show()
```
